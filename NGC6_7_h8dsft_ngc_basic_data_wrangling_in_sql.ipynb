```python
# Import Libraries
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()

client = bigquery.Client(project="hacktiv8-438003")
```

1. How many tables are in the austin_bikeshare dataset?

```python
client.query('''
SELECT COUNT(DISTINCT table_name) AS jumlah_table
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`;
''').to_dataframe()
```

2. How many columns in the bikeshare_stations table have STRING data?

```python
client.query('''
SELECT COUNT(column_name) AS string_column
FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS
WHERE table_name = "bikeshare_trips" AND data_type = 'STRING';
''').to_dataframe()
```

3. Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes

```python
client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL AND duration_minutes > 1000;
''').to_dataframe()
```

4. Filter trips where the subscriber_type is "Walk Up" and the end_station_name  is "Stolen".5


```python
client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE subscriber_type = 'Walk Up' AND end_station_name = 'Stolen';
''').to_dataframe()
```

```python
client.query('''
SELECT
    CAST(LEFT(Filing_Date, 4) AS INT) AS year,
    CAST(SUBSTRING(Filing_Date, 5, 2) AS INT) as month,
    COUNT(*) AS total_patent
FROM `patents-public-data.uspto_oce_cancer.publications`
WHERE CAST(LEFT(filing_date, 4) AS INT) BETWEEN 2012 AND 2023
  AND patent_title LIKE '%data%'
GROUP BY year, month
ORDER BY year DESC, month ASC;
''').to_dataframe()
```
