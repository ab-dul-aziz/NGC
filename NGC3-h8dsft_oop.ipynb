```python
class Book:
    def __init__(self, identifier, title, author):
        self.identifier = identifier
        self.title = title
        self.author = author
    
    def __str__(self):
        return f'ID: {self.identifier}, Title: {self.title}, Author: {self.author}'

class LibraryCatalog:
    def __init__(self):
        self.catalog = {}

    def add_book(self, book):
        """Adds a book to the catalog"""
        if book.identifier in self.catalog:
            print(f"Book with ID {book.identifier} already exists.")
        else:
            self.catalog[book.identifier] = book
            print(f"Book '{book.title}' by {book.author} added to the catalog.")
            

    def search_by_title(self, title):
        """Searches for books by title"""
        results = [book for book in self.catalog.values() if title.lower() in book.title.lower()]
        if results:
            print(f"Books found with title '{title}':")
            for book in results:
                print(book)
        else:
            print(f"No books found with title '{title}'.")

    def search_by_author(self, author):
        """Searches for books by author"""
        results = [book for book in self.catalog.values() if author.lower() in book.author.lower()]
        if results:
            print(f"Books found by author '{author}':")
            for book in results:
                print(book)
        else:
            print(f"No books found by author '{author}'.")

    def remove_book(self, identifier):
        """Removes a book from the catalog by its unique identifier"""
        try:
            removed_book = self.catalog.pop(identifier)
            print(f"Book '{removed_book.title}' by {removed_book.author} removed from the catalog.")
        except KeyError:
            print(f"Error: No book found with ID {identifier}.")

    def display_books(self):
        """Displays all the books in the catalog"""
        if not self.catalog:
            print("No books in the catalog.")
        else:
            print("Books in the catalog:")
            for book in self.catalog.values():
                print(book)

                
```

```python
class Book:
    def __init__(self, identifier, title, author):
        self.identifier = identifier
        self.title = title
        self.author = author
    
    def __str__(self):
        return f'ID: {self.identifier}, Title: {self.title}, Author: {self.author}'


class LibraryCatalog:
    def __init__(self):
        self.catalog = {}

    def add_book(self):
        """Adds a book to the catalog by taking input from the user"""
        try:
            identifier = int(input("Enter Book ID: "))
            title = input("Enter Book Title: ")
            author = input("Enter Book Author: ")
            if identifier in self.catalog:
                print(f"Book with ID {identifier} already exists.")
            else:
                book = Book(identifier, title, author)
                self.catalog[identifier] = book
                print(f"Book '{title}' by {author} added to the catalog.\n")
                print(self.catalog)
        except ValueError:
            print("Invalid input! Please enter a numeric value for Book ID.\n")

    def search_by_title(self):
        """Searches for books by title"""
        title = input("Enter the title to search: ")
        results = [book for book in self.catalog.values() if title.lower() in book.title.lower()]
        if results:
            print(f"Books found with title '{title}':")
            for book in results:
                print(book)
        else:
            print(f"No books found with title '{title}'.")

    def search_by_author(self):
        """Searches for books by author"""
        author = input("Enter the author to search: ")
        results = [book for book in self.catalog.values() if author.lower() in book.author.lower()]
        if results:
            print(f"Books found by author '{author}':")
            for book in results:
                print(book)
        else:
            print(f"No books found by author '{author}'.")

    def remove_book(self):
        """Removes a book from the catalog by its unique identifier"""
        try:
            identifier = int(input("Enter the Book ID to remove: "))
            removed_book = self.catalog.pop(identifier)
            print(f"Book '{removed_book.title}' by {removed_book.author} removed from the catalog.\n")
        except KeyError:
            print(f"No book found with ID {identifier}.\n")
        except ValueError:
            print("Invalid input! Please enter a numeric value for Book ID.\n")

    def display_books(self):
        """Displays all the books in the catalog"""
        print(self.catalog)
        if not self.catalog:
            print("No books in the catalog.")
        else:
            print("Books in the catalog:")
            for book in self.catalog.values():
                print(book)

    def ui_interface(self):
        """User Interface to interact with the library catalog"""
        while True:
            print("\nLibrary Catalog System")
            print("1. Add Book")
            print("2. Search Book by Title")
            print("3. Search Book by Author")
            print("4. Remove Book")
            print("5. Display All Books")
            print("6. Exit")
            
            try:
                choice = int(input("Choose an option (1-6): "))
                if choice == 1:
                    self.add_book()
                elif choice == 2:
                    self.search_by_title()
                elif choice == 3:
                    self.search_by_author()
                elif choice == 4:
                    self.remove_book()
                elif choice == 5:
                    self.display_books()
                elif choice == 6:
                    print("Exiting the system. Goodbye!")
                    break
                else:
                    print("Invalid choice! Please select a valid option.")
            except ValueError:
                print("Invalid input! Please enter a number between 1 and 6.\n")


# Example usage:
if __name__ == "__main__":
    library = LibraryCatalog()
    library.ui_interface()
```
